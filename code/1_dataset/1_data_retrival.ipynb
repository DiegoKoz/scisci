```python
def get_article_info(identifier):
    try:
        article_info = AbstractRetrieval(identifier=identifier,view='FULL')
        dict_info = {'title': article_info.title,
                     'abstract': article_info.abstract,
                     'affiliation': article_info.affiliation,
                     'aggregationType': article_info.aggregationType,
                     'authkeywords': article_info.authkeywords,
                     'authorgroup': article_info.authorgroup,
                     'authors': article_info.authors,
                     'chemicals': article_info.chemicals,
                     'citedby_count': article_info.citedby_count,
                     'coverDate': article_info.coverDate,
                     'description': article_info.description,
                     'doi': article_info.doi,
                     'funding': article_info.funding,
                     'funding_text': article_info.funding_text,
                     'citedby_link': article_info.citedby_link,
                     'contributor_group': article_info.contributor_group,
                     'language': article_info.language,
                     'publicationName': article_info.publicationName,
                     'references': article_info.references,
                     'subject_areas': article_info.subject_areas,
                     'issueIdentifier': article_info.issueIdentifier,
                     'volume': article_info.volume,
                     'eid': article_info.eid}
        logger.info('Info from: {} recived'.format(identifier))
    except Scopus429Error:
        # Remove the last item in _keys to assign it as new API key
        config["Authentication"]["APIKey"] = _keys.pop()
        logger.warning('changed API key')
        article_info = AbstractRetrieval(identifier=identifier,view='FULL')
        dict_info = {'title': article_info.title,
                     'abstract': article_info.abstract,
                     'affiliation': article_info.affiliation,
                     'aggregationType': article_info.aggregationType,
                     'authkeywords': article_info.authkeywords,
                     'authorgroup': article_info.authorgroup,
                     'authors': article_info.authors,
                     'chemicals': article_info.chemicals,
                     'citedby_count': article_info.citedby_count,
                     'coverDate': article_info.coverDate,
                     'description': article_info.description,
                     'doi': article_info.doi,
                     'funding': article_info.funding,
                     'funding_text': article_info.funding_text,
                     'citedby_link': article_info.citedby_link,
                     'contributor_group': article_info.contributor_group,
                     'language': article_info.language,
                     'publicationName': article_info.publicationName,
                     'references': article_info.references,
                     'subject_areas': article_info.subject_areas,
                     'issueIdentifier': article_info.issueIdentifier,
                     'volume': article_info.volume,
                     'eid': article_info.eid}
        logger.info('Info from: {} recived'.format(identifier))
    except Exception as e:
        logger.error('Failed at {}'.format(identifier))
        logger.error('Original message: {}'.format(e))
        dict_info = dict()
    finally:
        return dict_info
```
